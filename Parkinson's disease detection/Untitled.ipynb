## from above it is clear that 147 are diseases patients and 48 are non diseases. ##

```python
X = df.drop(columns = ['name','status'], axis = 1)
y = df['status']
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.2, stratify = y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn import svm
```

```python
model3 = svm.SVC(kernel = 'linear')
```

```python
model3.fit(X_train, y_train)
```

## Model Evaluation & Accuracy score . ##

```python
from sklearn.metrics import accuracy_score
```

```python
X_train_predicted = model3.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predicted, y_train)
```

```python
print("Accuracy of the training data is : ", training_data_accuracy)
```

```python
##Now the accuracy prediction of test data. ##
```

```python
X_test_predicted = model3.predict(X_test)
training_data_accuracy = accuracy_score(X_test_predicted, y_test)
```

```python
print("Accuracy of the test data is : ", training_data_accuracy)
```
